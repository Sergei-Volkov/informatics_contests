 # Имеется некоторая упорядоченная последовательность размера N неповторяющихся натуральных чисел. Требуется проверить наличие числа K в ней и вывести его порядковый номер. Если такого числа нет - вывести -1.

```python
size = int(input())  # size of sorted sequence
number = int(input())  # desired number
sequence = list( map( int, input().split() ) )  # sorted sequence


def bin_search(array, target, size):
        '''
            Binary search in sorted sequence.
            
            Input:
            
            array - sorted list
            target - element we're searching for
            size - length of the list
            
            Output:
            
            Position of element in array (starting from 1). 
            If there's no such element in array, function returns -1.
        '''
        left, right = 0, size - 1
        
        while left <= right:  # checking numbers from array
            middle = (left + right) // 2  # pivot
            if array[middle] < target:  # go to right half of new array
                left = middle + 1   
            elif array[middle] > target:  # go to left half of new array
                right = middle - 1 
            else:
                return middle + 1  # we found target number
        else:
            return -1  # no target number in array
        

print(bin_search(sequence, number, size))
```

 # Имеется некоторая упорядоченная последовательность неповторяющихся натуральных чисел. Требуется проверить наличие числа K в ней и вывести его порядковый номер (нумерация с единицы) с помощью рекурсии. Если такого числа нет - вывести -1. Циклы запрещены!

```python
sequence = list( map( int, input().split() ) )  # sorted sequence
number = int(input())  # desired number


def bin_search_recursive(array, target, left, right):
        '''
            Recursive bnary search in sorted sequence.
            
            Input:
            
            array - sorted list
            target - element we're searching for
            size - length of the list
            
            Output:
            
            Position of element in array (starting from 1). 
            If there's no such element in array, function returns -1.
        '''
        if left <= right:  # checking array both from left and right
            middle = (left + right) // 2  # pivot
            if array[middle] < target:  # go to right half of new array
                return bin_search_recursive(array, target, left + 1, right)
            elif array[middle] > target:  # go to left half of new array
                return bin_search_recursive(array, target, left, right - 1)
            else:  # we found target number
                return middle + 1
        else:  # no target number in array
            return -1
        
        
print(bin_search_recursive(sequence, number, 0, len(sequence) - 1))
```

 # Реализовать функцию find_root(f, a, b, tol) , реализующую поиск на интервале (a,b) корня x0 уравнения f(x) = 0 методом бисекции. Точность определения x0 не должна превышать значение tol . Выполнение равенства f = 0 проверять с точностью eps = 1e-6 . Гарантируется наличие единственного корня на рассматриваемом интервале. Функция f непрерывна и имеет значения разных знаков на концах интервала (a, b).

```python
eps = 1e-6  # root localization accuracy

def find_root(f, a, b, tol):
    '''
        Search for root of f(x) = 0 on interval (a, b) 
        with localization accuracy tol using bisection.
        
        Input:
        
        f - function, f(a) * f(b) < 0
        a, b - float, ends of segment
        tol - float, localization accuracy
        
        Output:
        
        x0 - solution of f(x) = 0 on (a, b)
    '''
    mid = (a + b) / 2  # pivot
    
    if f(mid) == 0:  # if f(pivot) = 0, then we have found x0
        return(mid)
    else:
         # while accuracy isn't reached yet, do
        while abs( f(mid) ) > eps and abs(a - mid) > tol:
            if f(mid) * f(a) > 0:  # go to right half
                a = mid
            elif f(mid) * f(b) > 0:  # go to left half
                b = mid
            mid = (a + b) / 2  # change pivot
        else:
            return(mid)  # return x0 with some accuracy
```

 # Однажды Шрек решил сходить в казино. Так как он там был впервые, он решил сыграть в довольно простую игру.

На игровом столе лежат N карточек. На каждой карточке написано целое положительное число. Игра проходит между игроком и крупье. Карточки лежат на столе числами вниз. Игра заключается в том, что игрок открывает ровно N/2 карточек. Сумма всех чисел, написанных на карточках, открытых игроком, называется суммой игрока». Следующим ходом крупье открывает оставшиеся N/2 карточек. Сумма всех чисел, написанных на карточках открытых крупье, называется «суммой крупье». Выигрыш игрока определяется разностью чисел между «суммой игрока» и «суммой крупье». Очевидно, что полученная разность может быть отрицательным числом. Это свидетельствует о том, что игрок проиграл и должен казино соответствующую сумму.

Все бы ничего, но Шрек обладает способностью видеть надписи сквозь бумагу любой плотности. Ваша задача определить максимальную сумму выигрыша, которую может получить Шрек с учётом того, что он видит все числа, написанные на карточках.

Примечание: встроенные сортировки использовать запрещено!

```python
amount = int(input())  # amount of cards
cards = list( map( int, input().split() ) )  # values on cards


def merge(L, R):
    '''
        Joining 2 sorted by increase lists.
        
        Input:
        
        L, R - arrays, sorted by increase (Left and Right).
        
        Output:
        
        joined_array - sorted by increase array which is sum of input ones 
    '''
    joined_array = [0] * (len(L) + len(R))  # creating future joined array
    pivot_L, pivot_R, pivot_J = 0, 0, 0  # placing pivots in beggings of lists
    
    while pivot_L != len(L) and pivot_R != len(R):  # joining items until pivot_L or pivot_R isn't equal
                                                    # to length of appropriate list
        if L[pivot_L] <= R[pivot_R]:                           
            joined_array[pivot_J] = L[pivot_L]
            pivot_L += 1
            pivot_J += 1
        elif R[pivot_R] <= L[pivot_L]:
            joined_array[pivot_J] = R[pivot_R]
            pivot_R += 1
            pivot_J += 1

    if pivot_L != len(L):  # joining last items to joined array
        while pivot_J != len(joined_array):
            joined_array[pivot_J] = L[pivot_L]
            pivot_J += 1
            pivot_L += 1

    elif pivot_R != len(R):
        while pivot_J != len(joined_array):
            joined_array[pivot_J] = R[pivot_R]
            pivot_J += 1
            pivot_R += 1    
    
    return(joined_array)


def merge_sort(A, depth=1, part='left'):
    '''
        Merge sort of array by recursively dividing it 
        in two halfs.
        
        Input:
        
        A - list of numbers.
        depth - depth of recursion
        part - which part is under merging
        
        Output:
        
        All steps of merging.
    '''
    #print('depth:', depth, '|', 'part:', part, '|', 'array:', A)   # result before using function
    if len(A) > 1:  # dividing array if it's length is > 1
        left = A[: len(A) // 2]
        right = A[len(A) // 2 :]
        
        merge_sort(left, depth + 1)  # sorting left half
        merge_sort(right, depth + 1, 'right')  # same for right half
        
        A[:] = merge(left, right)  # joining two arrays
    
        #print('depth:', depth, '|', 'part:', part, '|', 'after merge:', A)   # printing result before
                                                                             # before quitting function
    return(A)


cards = merge_sort(cards)  # sorting our list of values

max_sum = sum(cards[int(amount / 2) :])  # sum of first half - min sum
min_sum = sum(cards[: int(amount / 2)])  # sum of last half - max sum
print(max_sum - min_sum)
```
