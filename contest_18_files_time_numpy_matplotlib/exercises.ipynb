## Скачайте входной файл task1.txt. В каждой его строчке содержится по несколько чисел через пробел. Вам необходимо создать файл out.txt, в каждой строчке которого содержится среднее арифметическое чисел из соответствующей строчки входного файла.

```python
import numpy as np

out = open('out.txt', 'w')

with open('task1.txt') as inp:
    for line in inp:
        print(np.mean(np.array(list(map(float, line.split())))),
              file=out
             )
out.close()
```

## Проверьте асимптотики следующих групп алгоритмов

алгоритмы сортировки: вставками, кучей, встроенная list.sort()

добавление элемента в начало: list.insert(x, 0) и deque.appendleft(), взятие элемента из начала: deque.popleft() и list.pop(0)

поиск элемента: операции elem in list и elem in set

Для каждой группы постройте график асимптотики - зависимости времени работы алгоритма от характерного масштаба задачи.

```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

### Сортировка вставками

```python
def insertion_sort(numbers):
    for i in range( 1, len(numbers) ):
        j = i - 1 ## «cutting» part of list before i-th element
        key = numbers[i] ## these elements are compared with the i-th element
        while numbers[j] > key and j >= 0: ## comparison: if element is greater than
            numbers[j + 1], numbers[j] = numbers[j], numbers[j+1] ## the one on the right, swap them
            j -= 1
        numbers[j + 1] = key ## next comparision is the greatest element
```

```python
times_ins = []

for size in np.arange(1, 5000, 50):
    lst = np.random.random(size=size)
    
    start = time.time()
    insertion_sort(lst)
    end = time.time()
    
    times_ins.append(end-start)
```

### Сортировка кучей

```python
def down(array, i, end):  # creating binary heap
    if end > 2*i + 2:  # if we have 2 children
        if array[2*i + 2] > array[2*i + 1]:  # if one children is bigger then other
            if array[2*i + 2] > array[i]:  # if children is bigger then parent
                array[2*i + 2], array[i] = array[i], array[2*i + 2]  # swap them
                down(array, 2*i + 2, end)  # next stage of pyramid
        else:  # if other children is bigger
            if array[2*i + 1] > array[i]:  # if children is bigger then parent
                array[2*i + 1], array[i] = array[i], array[2*i + 1]  # swap them
                down(array, 2*i + 1, end)  # next stage of pyramid
    elif end > 2*i + 1:  # if we have 1 child
        if array[2*i + 1] > array[i]:  # if child is bigger then parent
            array[2*i + 1], array[i] = array[i], array[2*i + 1]  # swap them
            down(array, 2*i + 1, end)  # next stage
    
def heapsort(abc):
    for i in range(len(abc) // 2 + 1, -1, -1):  # create binary heap
        down(abc, i, len(abc))
    
    for i in range(len(abc) - 1):  # cut last (biggest) element and iterate throw lasting part
        abc[0], abc[- i - 1] = abc[- i - 1], abc[0]
        down(abc, 0, len(abc) - i - 1)
```

```python
times_heap = []

for size in np.arange(1, 10000, 50):
    lst = np.random.random(size=size)
    
    start = time.time()
    heapsort(lst)
    end = time.time()
    
    times_heap.append(end-start)
```

### Метод list.sort()

```python
times_sort = []

for size in np.arange(1, 100000, 50):
    lst = np.random.random(size=size)
    
    start = time.time()
    lst.sort()
    end = time.time()
    
    times_sort.append(end-start)
```

```python
fig = plt.figure(figsize=(16, 8))
fig.add_subplot(131)
plt.plot(np.arange(1, 5000, 50), times_ins, label='Insertion Sort')
plt.xlabel('Характерный размер задачи (размер списка)')
plt.ylabel('Время выполнения сортировки, с')
plt.legend()
plt.grid()
fig.add_subplot(132)
plt.plot(np.arange(1, 10000, 50), times_heap, label='Heap Sort')
plt.xlabel('Характерный размер задачи (размер списка)')
plt.ylabel('Время выполнения сортировки, с')
plt.title('Sorts')
plt.legend()
plt.grid()
fig.add_subplot(133)
plt.scatter(np.arange(1, 100000, 50), times_sort, marker='.', alpha=0.5, label='list.sort() Method')
plt.xlabel('Характерный размер задачи (размер списка)')
plt.ylabel('Время выполнения сортировки, с')
plt.legend()
plt.grid()
plt.show()
```

### Функции добавления элемента в список

```python
import collections
```

```python
# list.insert()
times_insert = []
lst = [list(np.random.random(i)) for i in np.arange(0, 10000, 50)]
for i in range(len(lst)):
    start = time.time()
    lst[i].insert(np.random.randint(500), 0)
    end = time.time()
    times_insert.append(end-start)
```

```python
# list.pop(0)
times_pop = []
lst = [list(np.random.random(i)) for i in np.arange(1, 10000, 50)]
for i in range(len(lst)):
    start = time.time()
    lst[i].pop(0)
    end = time.time()
    times_pop.append(end-start)
```

```python
# deque.appendleft()
times_q_app = []
de = [collections.deque(np.random.random(i)) for i in np.arange(0, 10000, 50)]
for i in range(len(de)):
    start = time.time()
    de[i].appendleft(np.random.randint(500))
    end = time.time()
    times_q_app.append(end-start)
```

```python
# deque.popleft()
times_q_pop = []
de = [collections.deque(np.random.random(i)) for i in np.arange(1, 10000, 50)]
for i in range(len(de)):
    start = time.time()
    de[i].popleft()
    end = time.time()
    times_q_pop.append(end-start)
```

```python
plt.figure(figsize=(16, 8))
plt.scatter(np.arange(0, 10000, 50), times_insert, marker='o', alpha=0.5, label='list.insert() Method')
plt.scatter(np.arange(0, 10000, 50), times_pop, marker='o', alpha=0.5, label='list.pop() Method')
plt.scatter(np.arange(0, 10000, 50), times_q_app, marker='o', alpha=0.5, label='deque.appendleft() Method')
plt.scatter(np.arange(0, 10000, 50), times_q_pop, marker='o', alpha=0.5, label='deque.popleft() Method')
plt.xlabel('Характерный размер задачи (размер списка)')
plt.ylabel('Время выполнения вставки, с')
plt.title('Inserts & Pops')
plt.legend()
plt.grid()
plt.show()
```

### Поиск элемента в list и в set

```python
# list
times_l = []
lst = [list(np.random.randint(1, 10000, size=i)) for i in np.arange(1, 50000, 50)]
for i in range(len(lst)):
    start = time.time()
    np.random.randint(1, 10000) in lst[i]
    end = time.time()
    times_l.append(end-start)
```

```python
# set
times_s = []
s = [set(np.random.randint(1, 10000, size=i)) for i in np.arange(1, 50000, 50)]
for i in range(len(lst)):
    start = time.time()
    np.random.randint(1, 10000) in s[i]
    end = time.time()
    times_s.append(end-start)
```

```python
plt.figure(figsize=(16, 8))
plt.scatter(np.arange(0, 50000, 50), times_l, marker='o', alpha=0.5, label='In list')
plt.scatter(np.arange(0, 50000, 50), times_s, marker='o', alpha=0.5, label='In set')
plt.xlabel('Характерный размер задачи (размер списка)')
plt.ylabel('Время нахождения элемента, с')
plt.title('Element search')
plt.legend()
plt.grid()
plt.show()
```
